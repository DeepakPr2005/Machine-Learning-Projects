```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


x = [1,2,3,4,5,6,7,8,9]
y = [10,20,30,40,50,60,70,80,55]
plt.title("Deepak kumar Age Analysis")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.plot(x,y , color= "g", linestyle =":",linewidth=3)
plt.show()
```

```python
x = np.random.randint(10,50,8)
x1 = 2*x
x2 = 9*x
plt.plot(x,x1 , color= "g", linestyle =":",linewidth=3)
plt.plot(x,x2 , color= "g", linestyle =":",linewidth=3)
plt.title("Deepak kumar Age Analysis")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.grid(True)
plt.show()
```
