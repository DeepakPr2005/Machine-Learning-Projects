```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("iris[1].csv")
iris.head(5)
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"])

plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.show()
```

```python
iris["Speciess"]  = iris["Species"].replace({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2})
iris.head()
```

```python
iris.shape
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Speciess"])

plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.show()
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Speciess"],cmap="winter")

plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.show()
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Speciess"])

plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.colorbar()
plt.show()
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Speciess"],alpha=0.5)
# alpha is show the instensity of the color
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Speciess"],alpha=0.5)
# alpha is show the instensity of the color
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.colorbar()
plt.show()
```

# Annotation

```python
x=  [1,2,3,4]
y=  [5,6,7,8]

plt.scatter(x,y)

plt.text(1,5,"Point 1")
plt.text(2,6,"Point 2")
plt.text(3,7,"Point 3")
plt.text(4,8,"Point 4")

plt.show()
```

```python
x=  [1,2,3,4]
y=  [5,6,7,8]

plt.scatter(x,y)

plt.text(1,5,"Point 1",fontdict={"size":10,"color":"green"})
plt.text(2,6,"Point 2",fontdict={"size":14,"color":"black"})
plt.text(3,7,"Point 3",fontdict={"size":18,"color":"gray"})
plt.text(4,8,"Point 4",fontdict={"size":22,"color":"red"})

plt.show()
```
