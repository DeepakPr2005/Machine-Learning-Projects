# MEARGING and CONCAT

```python
import pandas as pd
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,5],"C":[60,70,80,90,100]})

pd.merge(a1,a2)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,5],"C":[60,70,80,90,100]})

pd.merge(a2,a1)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a2,a1)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a1,a2,how="inner")
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a1,a2,how="left")
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a1,a2,how="right")
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a1,a2,how="left",indicator=True)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[60,70,80,90,100]})

pd.merge(a1,a2,how="outer",indicator=True)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[60,70,80,90,100]})

pd.merge(a1,a2,left_index=True,right_index=True)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
a2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[60,70,80,90,100]})

pd.merge(a1,a2,left_index=True,right_index=True,suffixes=("_Name","_ID"))
```

#        ----------Goupby-------

# -----------------------------Pivot----------------------------

```python
d = pd.DataFrame({  "days":[1,2,3,4,5],
                    "Name":["a","b","c","d","d"],
                    "S_1":[10,20,30,40,50],
                    "S_2":[60,70,80,78,85]})

d
```

```python
d.pivot(index="days",columns="Name")
```

```python
d.pivot_table(columns="days",index="Name",aggfunc="mean")
```

```python
d.pivot_table(columns="days",index="Name",aggfunc="sum")
```

```python
d.pivot_table(columns="days",index="Name",aggfunc="sum",margins=True)
```
