```python
col_list = []

for col in flight_book.columns:
    if((flight_book[col].dtype != 'object') & (col != "price")):
        col_list.append(col)

x= flight_book[col_list]
vif_data = pd.DataFrame()
vif_data["Feature"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]

print(vif_data)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = flight_book.drop(columns=['price','flight'])

y = flight_book['price']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

linear_model = LinearRegression()

linear_model.fit(x_train,y_train)

```

```python
y_pred = linear_model.predict(x_test)

differance = pd.DataFrame(np.c_[y_test,y_pred],columns=["Actual_Value","Predicted_value"])
differance.head()
```

```python
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.metrics import mean_absolute_percentage_error

print("R2 Score",r2_score(y_test,y_pred))

mean_abs_error = metrics.mean_absolute_error(y_test,y_pred)
print("mean absolute error",mean_abs_error)

print("mean absolute percentage error",mean_absolute_percentage_error(y_test,y_pred))

mean_sq_error = metrics.mean_squared_error(y_test,y_pred)
print("mean squared error",mean_sq_error)

root_mean_squared_error = np.sqrt(mean_sq_error)
print("root mean squared error",root_mean_squared_error)
```

```python
# using histogram to visualize the difference between actual and predicted prices
sns.histplot(y_test,label="Actual",color="blue")
sns.histplot(y_pred,label="Predicted",color="grey")
plt.title("Lenear Regression Model")
plt.show()
```
