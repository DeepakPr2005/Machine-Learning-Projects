```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_df = batter.head(150).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df["Avg"],sample_df["Strike_Rate"])
plt.show()
```

```python
plt.scatter(sample_df["Avg"],sample_df["Strike_Rate"])

for i in range(sample_df.shape[0]):
    plt.text(sample_df["Avg"].values[i],sample_df["Strike_Rate"].values[i],sample_df["Better"].values[i])

plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df["Avg"],sample_df["Strike_Rate"])

for i in range(sample_df.shape[0]):
    plt.text(sample_df["Avg"].values[i],sample_df["Strike_Rate"].values[i],sample_df["Better"].values[i])

plt.show()
```

```python
'''We wants to show the size of the each dot
based on the runs scored by the bastmen'''


plt.figure(figsize=(15,10))
plt.scatter(sample_df["Avg"],sample_df["Strike_Rate"],s=sample_df["Runs"])

for i in range(sample_df.shape[0]):
    plt.text(sample_df["Avg"].values[i],sample_df["Strike_Rate"].values[i],sample_df["Better"].values[i])

plt.show()
```

```python
# suppose we wants batsman who has strike rate of 130 for that we can plot a line

plt.figure(figsize=(18,10))
plt.scatter(sample_df["Avg"],sample_df["Strike_Rate"])


plt.axhline(140,color="red")
plt.axhline(100,color="black")
plt.axvline(15,color="blue")


for i in range(sample_df.shape[0]):
    plt.text(sample_df["Avg"].values[i],sample_df["Strike_Rate"].values[i],sample_df["Better"].values[i])

plt.show()
```

# 3D surface plot

```python
x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100)
```

```python
xx,yy= np.meshgrid(x,y)
xx.shape
```

```python
# we wants to create z with help of x and y
z = xx**2+yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection="3d")

ax.scatter3D(xx,yy,z)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection="3d")

p= ax.plot_surface(xx,yy,z,cmap="viridis")

fig.colorbar(p)
plt.show()
```

```python
x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100)


xx,yy= np.meshgrid(x,y)
# xx.shape

z = np.sin(xx)+np.cos(yy)

ax = plt.subplot(projection="3d")

p= ax.plot_surface(xx,yy,z,cmap="viridis")

fig.colorbar(p)
plt.show()
```

# Contour Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
# filled cointour graph
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```
